```python
import gymnasium as gym

if __name__ == "__main__":
  env = gym.make("CartPole-v1")
  total_reward = 0
  total_steps = 0
  obs, _ = env.reset()
while True:
  action = env.action_space.sample()
  obs,reward,is_done,is_trunc,_ = env.step(action)
  total_reward+=reward
  total_steps += 1
  if is_done==True:
    break
print(total_reward,"rewards in ",total_steps, "steps")
```
