```python
# testing the environment
import gymnasium as gym

if __name__ == "__main__":
  env = gym.make("CartPole-v1")
  total_reward = 0
  total_steps = 0
  obs, _ = env.reset()
while True:
  action = env.action_space.sample()
  obs,reward,is_done,is_trunc,_ = env.step(action)
  total_reward+=reward
  total_steps += 1
  if is_done==True:
    break
print(total_reward,"rewards in ",total_steps, "steps")
```

```python
import random
from gymnasium.wrappers import RecordVideo
from IPython.display import Video
import glob
class RandomActionWrapper(gym.ActionWrapper):
  def __init__(self, env: gym.Env, epsilon: float= 0.5):
    super(RandomActionWrapper,self).__init__(env)
    self.epsilon = epsilon
  def action(self, action: gym.core.WrapperActType):
    if random.random()<self.epsilon:
       action = self.env.action_space.sample()
       print(action)
    return action
if __name__ == "__main__":
  env = RandomActionWrapper(gym.make("CartPole-v1", render_mode = "rgb_array"))
  env = RecordVideo(env,video_folder="videos",episode_trigger=lambda episode_id: True)

  obs,_ = env.reset()
  total_reward = 0

  while True:
    obs,reward,done,_,_ = env.step(0)
    total_reward += reward
    if done:
      break
  print(total_reward)
env.close()


video_path = glob.glob("videos/*.mp4")[0]
Video(video_path, embed=True)
```
